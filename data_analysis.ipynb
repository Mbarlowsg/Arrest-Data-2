```python
with open('datasets/arrest_data.geojson') as data_file:    
    d= json.load(data_file)  

df = pd.json_normalize(d['features'])

df.head()
```

```python
precincts = {}

for feature in d['features']:
    arrest_precinct = feature['properties']['arrest_precinct']
    precincts[arrest_precinct] = {}
```

```python
for feature in d['features']:
    arrest_precinct = feature['properties']['arrest_precinct']
    perp_race = str(feature['properties']['perp_race'])
    perp_sex = str(feature['properties']['perp_sex'])

    if 'precinct_total_arrests' in precincts[arrest_precinct].keys():
        precincts[arrest_precinct]['precinct_total_arrests'] += 1   
    else:
        precincts[arrest_precinct]['precinct_total_arrests'] = 1 
    
    if 'precinct_perp_race' in precincts[arrest_precinct].keys():
        if perp_race in precincts[arrest_precinct]['precinct_perp_race'].keys():
            precincts[arrest_precinct]['precinct_perp_race'][perp_race] += 1
        else:
            precincts[arrest_precinct]['precinct_perp_race'][perp_race] = 1
    else:
        precincts[arrest_precinct]['precinct_perp_race'] = {perp_race: 1}

    if 'precinct_perp_sex' in precincts[arrest_precinct].keys():
        if perp_sex in precincts[arrest_precinct]['precinct_perp_sex']:
            precincts[arrest_precinct]['precinct_perp_sex'][perp_sex] += 1
        else:
            precincts[arrest_precinct]['precinct_perp_sex'][perp_sex] = 1
    else:
        precincts[arrest_precinct]['precinct_perp_sex'] = {perp_sex: 1}
```

```python
json_object = json.dumps(precincts, indent=4)

with open('precinct_stats.js', 'w') as writer:
    # writer.writelines(object)
    # json.dump(object, writer, ensure_ascii=True, indent=4, allow_nan=True)
    writer.write(json_object)
```
